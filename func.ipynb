```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return np.maximum(0, x)

def swish(x):
  return x * (1 / (1 + np.exp(-x)))

x = np.linspace(-5, 5, 100)

plt.plot(x, relu(x), label="ReLU")
plt.plot(x, swish(x), label="SiLU")

plt.xlabel("x")
plt.ylabel("y")
# plt.title("ReLU vs Swish")
plt.legend()
plt.grid(True)
plt.savefig('relu_silu.png', dpi=400)

plt.show()
```
